```python
import matplotlib.pyplot as plt
```

```python
# Años para la proyeccion ilustrativa
anios_proj = [1970, 1980, 1990, 2000, 2010, 2020, 2030, 2040]
```

```python
# Valores para cada modelo
pobl_lineal = [2.08, 2.40, 2.73, 3.07, 3.31, 3.84, 4.43, 5.10]       # Crecimiento constante
pobl_logistico = [2.08, 2.40, 2.73, 3.07, 3.31, 3.84, 4.60, 4.90]    # Con desaceleración
pobl_exponencial = [2.08, 2.40, 2.73, 3.07, 3.31, 3.84, 4.81, 6.10]  # Crecimiento acelerado
```

```python

# grafico
plt.figure(figsize=(10, 6))
plt.plot(anios_proj, pobl_lineal, label="Lineal", color='blue', marker='o')
plt.plot(anios_proj, pobl_logistico, label="Logístico", color='green', marker='o')
plt.plot(anios_proj, pobl_exponencial, label="Exponencial", color='red', marker='o')


plt.title("Grafico 2: Proyecciones extendidas de poblacion")
plt.xlabel("Año")
plt.ylabel("Población (millones)")
plt.xticks(anios_proj)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
